```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file (replace 'your_file.csv' with the path to your file)
data = pd.read_csv('input.csv')

# Extract age column
ages = data['age']

# Shapiro-Wilk test for normality
stat, p_value = stats.shapiro(ages)

# Print results of Shapiro-Wilk test
print(f"Shapiro-Wilk Test Statistic: {stat}")
print(f"p-value: {p_value}")

# Check normality based on p-value (common threshold p < 0.05)
if p_value > 0.05:
    print("Data is likely normally distributed (fail to reject the null hypothesis).")
else:
    print("Data is likely not normally distributed (reject the null hypothesis).")

# Plot histogram to visualize age distribution
plt.figure(figsize=(8, 5))
sns.histplot(ages, kde=True, bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Q-Q plot to further check normality visually
plt.figure(figsize=(8, 5))
stats.probplot(ages, dist="norm", plot=plt)
plt.title('Q-Q Plot for Age')
plt.show()
```
