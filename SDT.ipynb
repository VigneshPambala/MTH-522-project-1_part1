```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Path to the CSV file
file_path = 'input.csv'

# Load the data
df = pd.read_csv(file_path)

# Extract ages for Black and White individuals
ab = df[df['race'] == 'B']['age']
aw = df[df['race'] == 'W']['age']

# Calculate mean and standard deviation
mb = np.mean(ab)
sb = np.std(ab, ddof=1)
mw = np.mean(aw)
sw = np.std(aw, ddof=1)

# Print mean and standard deviation
print(f"Mean age of Black individuals: {mb:.2f}")
print(f"Mean age of White individuals: {mw:.2f}")
print(f"Standard deviation of age for Black individuals: {sb:.2f}")
print(f"Standard deviation of age for White individuals: {sw:.2f}")

# Calculate age difference
age_diff = np.abs(np.mean(ab) - np.mean(aw))
print(f"The average age difference between B and W is: {age_diff:.2f} years")

# Plot age distributions
plt.figure(figsize=(10, 6))
sns.histplot(ab, color='blue', kde=True, label='Black Individuals', bins=20, alpha=0.5)
sns.histplot(aw, color='orange', kde=True, label='White Individuals', bins=20, alpha=0.5)
plt.title('Age Distribution of Black and White Individuals')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```
